```python
import numpy as np
import pandas as pd
from statsmodels.tsa.api import VAR
from statsmodels.tsa.stattools import adfuller
```

```python
df1 = pd.read_csv('Data/Sparkling.csv')

df1['YearMonth'] = pd.to_datetime(df1['YearMonth'],
                                 format='%Y-%m')
df1 = df1.set_index('YearMonth')
df1 = df1.asfreq('MS')
df1.head()
```

```python
df2 = pd.read_csv('Data/Rose.csv', parse_dates=True)
df2['YearMonth'] = pd.to_datetime(df2['YearMonth'],
                                 format='%Y-%m')
df2 = df2.set_index('YearMonth')
df2 = df2.asfreq('MS')

df2.head()
```

```python
df2.isnull().sum()
```

```python
df2 = df2.interpolate(method='spline', order=1)
```

```python
joint_df = pd.merge(df2, df1, left_index=True, right_index=True)
joint_df.columns = ['Rose', 'Sparkling']
```

```python
# perform 1st differencing on the datset
joint_df_diff = joint_df.diff().dropna()
```

```python
# Step 4: Fit VAR model
model = VAR(joint_df_diff)
lag_order = model.select_order(maxlags=12)
print("Selected Lags:", lag_order.summary())

selected_lag = lag_order.selected_orders['aic']  
var_model = model.fit(selected_lag)
print(var_model.summary())
```

### VAR Model Interpretation — Key Bullets
- Optimal lag order = 11, as selected by AIC, BIC, HQIC, and FPE
- Model uses the past 11 months of data to forecast both wine types.

### Equation 1: Predicting Rosé Sales
- Strong negative autocorrelation:
    - Lags like L1.Rose = -0.69 (p < 0.001), L3.Rose = -0.77 (p < 0.001)
- Sparkling wine influences Rosé:
    - All 11 Sparkling lags are significant, e.g., L10.Sparkling = -0.0123 (p < 0.001),  L11.Sparkling = -0.0130 (p < 0.001)

➤ **Conclusion**: Rosé sales are influenced by both its own past and past Sparkling sales.

### Equation 2: Predicting Sparkling Sales
- Dominated by own lags:
  - Lags like L1.Sparkling = -0.97 (p < 0.001), L11.Sparkling = -0.94 (p < 0.001)
- Rosé lags are not significant:
  - All L#.Rose terms have p > 0.7, e.g., L4.Rose = -3.25 (p = 0.254)

### Conclusion: Sparkling sales are self-driven — Rosé has no measurable effect.
-  Cross-Dependency Summary
    - Directionality: Sparkling → Rosé (significant), but Rosé ✖ Sparkling (not significant)
    - Residual correlation = 0.36 → Moderate shared variation remains

```python
# Step 5: Forecast
forecast_input = joint_df_diff.values[-selected_lag:]
forecast = var_model.forecast(y=forecast_input, steps=12)
forecast_df = pd.DataFrame(forecast, columns=['Rose_forecast_diff', 'Sparkling_forecast_diff'])
```

```python
# Step 6: Convert forecast back to actual level
last_values = joint_df.iloc[-1]
forecast_df['Rose_forecast'] = forecast_df['Rose_forecast_diff'].cumsum() + last_values['Rose']
forecast_df['Sparkling_forecast'] = forecast_df['Sparkling_forecast_diff'].cumsum() + last_values['Sparkling']
```

```python
# Final forecast values
print(forecast_df[['Rose_forecast', 'Sparkling_forecast']])
```

### VAR Forecast Interpretation (Next 12 Months)
- The model forecasts both wine sales together, accounting for cross-influences
- Forecast values are in original scale, after converting from differenced output
____________________________________________________________________________________________________
### Rosé Wine Forecast (12 months ahead)
- Starts at: 39.3
- Peaks around: 95.5 (Month 5)
- Mostly fluctuates between 27–58 units/month

**Interpretation**:Rosé sales are expected to rise gradually with some mid-year peaks, potentially influenced by seasonal Sparkling effects.
____________________________________________________________________________________________________
### Sparkling Wine Forecast (12 months ahead)
- Starts at: 1,485 units
- Peaks at: 5,791 units (Month 5)
- Shows high volatility: drops to 1,241 (Month 6), rises again to 2,093 (Month 12)

**Interpretation**: Sparkling sales show more aggressive ups and downs, possibly due to strong seasonal cycles captured by autocorrelations.
____________________________________________________________________________________________________
### High-Level Summary Bullets
- The joint VAR model successfully generates 12-month forecasts for both wine types using mutual lagged dependencies.
- Sparkling drives Rosé, visible in the moderate correlation and significance of Sparkling lags in the Rosé equation.
- The model captures seasonal peaks and adjustments across both series — ideal for planning promotions or inventory.
- Forecasted values are now ready to be visualized, validated, or exported

```python
import matplotlib.pyplot as plt

# Create forecast timeline
forecast_index = pd.date_range(start=joint_df.index[-1] + pd.offsets.MonthBegin(1), periods=12, freq='MS')
forecast_df.index = forecast_index

# Plot Rosé forecast
plt.figure(figsize=(16, 6))
plt.plot(joint_df['Rose'], label='Historical Rose')
plt.plot(forecast_df['Rose_forecast'], label='Forecasted Rose', linestyle='--')
plt.title('12-Month Forecast for Rosé Wine')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.legend()
plt.grid(True)
plt.show()

# Plot Sparkling forecast
plt.figure(figsize=(16, 6))
plt.plot(joint_df['Sparkling'], label='Historical Sparkling')
plt.plot(forecast_df['Sparkling_forecast'], label='Forecasted Sparkling', linestyle='--')
plt.title('12-Month Forecast for Sparkling Wine')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.legend()
plt.grid(True)
plt.show()
```

## Add Confidence Intervals (Using Forecast Error Covariance)
VAR allows calculation of forecast error variance, which we use to build ±CI bands:

```python
# Get forecast and standard errors
forecast_res = var_model.forecast_interval(forecast_input, steps=12, alpha=0.05)

lower = pd.DataFrame(forecast_res[1], columns=['Rose_lower', 'Sparkling_lower'])
upper = pd.DataFrame(forecast_res[2], columns=['Rose_upper', 'Sparkling_upper'])

# Convert from diff to actual scale
lower['Rose_forecast'] = lower['Rose_lower'].cumsum() + last_values['Rose']
upper['Rose_forecast'] = upper['Rose_upper'].cumsum() + last_values['Rose']
lower['Sparkling_forecast'] = lower['Sparkling_lower'].cumsum() + last_values['Sparkling']
upper['Sparkling_forecast'] = upper['Sparkling_upper'].cumsum() + last_values['Sparkling']

# Assign forecast index
lower.index = forecast_index
upper.index = forecast_index

# Plot with confidence interval - Rosé
plt.figure(figsize=(16, 6))
plt.plot(joint_df['Rose'], label='Historical Rose')
plt.plot(forecast_df['Rose_forecast'], label='Forecasted Rose', linestyle='--')
plt.fill_between(forecast_index, lower['Rose_forecast'], upper['Rose_forecast'], color='gray', alpha=0.3, label='95% CI')
plt.title('Rosé Wine Forecast with 95% Confidence Interval')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.legend()
plt.grid(True)
plt.show()

# Plot with confidence interval - Sparkling
plt.figure(figsize=(16, 6))
plt.plot(joint_df['Sparkling'], label='Historical Sparkling')
plt.plot(forecast_df['Sparkling_forecast'], label='Forecasted Sparkling', linestyle='--')
plt.fill_between(forecast_index, lower['Sparkling_forecast'], upper['Sparkling_forecast'], color='gray', alpha=0.3, label='95% CI')
plt.title('Sparkling Wine Forecast with 95% Confidence Interval')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.legend()
plt.grid(True)
plt.show()
```

# Residual Monitoring Analysis - SPARKLING

```python
import joblib
sparkling_sarima_model = joblib.load('sparkling_sarima_model.pkl')
```

```python
residuals = sparkling_sarima_model.resid
residuals
```

```python
## Residuals Over Time

plt.figure(figsize=(16, 5))
plt.plot(residuals)
plt.axhline(0, linestyle='--', color='gray')
plt.title("Residuals Over Time - Sparkling")
plt.xlabel("Date")
plt.ylabel("Residual (Actual - Predicted)")
plt.grid(True)
plt.show()
```

### Plot 1 – Raw Residuals Over Time
- **Early instability (1979–1981)**:
    - Large spikes in residuals exceeding ±3000 units
    - Indicates the model struggled to capture dynamics in the early phase, likely due to initialization or lack of prior data

- **Post-1981 behavior**:
    - Residuals mostly stay within ±1000
    - Residuals oscillate around zero, suggesting they behave like white noise (good fit)

**Takeaway**: After initial stabilization, the residuals do not exhibit systematic trends or autocorrelation, indicating a reasonable model fit.

```python
## Rolling Mean & Std Dev

rolling_mean = residuals.rolling(window=12).mean()
rolling_std = residuals.rolling(window=12).std()

plt.figure(figsize=(16,5))
plt.plot(residuals, label='Residuals')
plt.plot(rolling_mean, label='Rolling Mean (12m)')
plt.fill_between(residuals.index, rolling_mean - 2*rolling_std, rolling_mean + 2*rolling_std,
                 color='gray', alpha=0.3, label='±2 Std Dev')
plt.title("Rolling Residual Diagnostics - Sparkling")
plt.legend()
plt.grid(True)
plt.show()
```

### Plot 2 – Rolling Diagnostics (Rolling Mean & Std Deviation)
- **Rolling Mean (Orange Line)**:
    - Remains close to zero throughout the time series
    - No signs of drift or long-term bias

- **±2 Standard Deviation Band (Gray Area)**:
    - Most residuals fall within the expected confidence bands
    - A few spikes (e.g., 1987 and 1994) occur but return to normal, showing the model recovers well

**Takeaway**: The residuals are relatively stable over time. The model errors are centered and have consistent variance, which supports its reliability.
